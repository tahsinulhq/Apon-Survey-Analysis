```python
def matching(str_1, num1, num2):

  counter = 0
  col4 = 'match'
  
  for i in range(1,num1,1):    
           str_2 = str(i)       
           col1= str_1 + str_2
          
           for j in range(num1,num2,1):
             str_3 = str(j)
             counter += 1
             count = str(counter)
             col2 = str_1 + str_3
             col5 = col4 + count

             df[col5] = df[col1] == df[col2]            
             df[col2] = np.where(df[col5] == True, np.NaN, df[col2] )

             df.drop(col5, axis =1, inplace = True)
```
